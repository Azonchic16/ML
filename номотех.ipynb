```python
def factorial(n):
    if n == 0:
        return 1

    return n * factorial(n-1)

def bc(n, k):
    return factorial(n) // (factorial(k) * factorial(n-k))

n1, n2, q1, q2, k, m = 6, 8, 0.2, 0.3, 2, 2

p_a = bc(n1,k) * q1 ** (2) * (1 - q1) ** (n1 - k)
p_b = p_a * bc(n2,m) * q2 ** (2) * (1 - q2) ** (n2 - m)
p_c = bc(n1,2) * q1 ** (2) * (1 - q1) ** (n1 - 2) * (1 - q2) ** (n2) + bc(n1,1) * q1 * (1 - q1) ** (n1 - 1) * bc(n2,1) * q2 * (1 - q2) ** (n2 - 1) + (1 - q1) ** (n1) * bc(n2, 2) * q2 ** (2) * (1 - q2) ** (n2 - 2)
p_d = 0
for i in range(1, n1 + 1):
    for j in range(0, i):
        p_d += bc(n1, i) * q1 ** (i) * (1 - q1) ** (n1 - i) * bc(n2, j) * q2 ** (j) * (1 - q2) ** (n2 - j)

p_e_c = (bc(n1,2) * q1 ** (2) * (1 - q1) ** (n1 - 2) * (1 - q2) ** (n2) + (1 - q1) ** (n1) * bc(n2, 2) * q2 ** (2) * (1 - q2) ** (n2 - 2)) / p_c
p_last = 0
for i in range(3, n1 + 1):
    for j in range(0, i):
        p_last += bc(n1,i) * q1 ** (i) * (1 - q1) ** (n1 - i) * bc(n2, j) * q2 ** (j) * (1 - q2) ** (n2 - j)
p_last = p_last / p_d

print(p_a, p_b, p_c, p_d, p_e_c, p_last)
```
