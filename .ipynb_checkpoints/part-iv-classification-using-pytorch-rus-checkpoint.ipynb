# Задача классификации

Задача классификации заключается в предсказании некоторой метки класса (из некоторого конечного множества) на основании набора признаков. Классифицировать объект — значит, указать номер (или наименование класса), к которому относится данный объект.

Примеры:

| Тип проблемы | Чем является? | Пример |
| :-----: | ----- | ----- |
| **Бинарная классификация** | Метка класса принимает два значения | Предсказание того, является ли письмо спамом или нет? Стоит ли выдавать кредит клиенту? |
| **Мультиклассовая классификация** | Метка класса может принимать больше двух значений | Определить стадию заболевания по симптомам. Определение какое из животных изображено на картинке? |
| **Классификация с пересекающимися классами** | Объект может одновременно относиться к нескольким классам | Предсказать основные проблемы, поднятые в статье |


# Архитектура нейронной сети для решения задачи классификации

| **Hyperparameter** | **Binary Classification** | **Multiclass classification** |
| :---: | --- | --- |
| **Входной слой** (`in_features`) | Содержит столько нейронов, сколькими признаками характеризуется объект (например, размер был бы равен 5, если человек характеризовался бы возрастом, полом, ростом, весом, курит человек или нет для решения задачи предсказания будет ли сердечный приступ в ближайшие 10 лет)  | Аналогично бинарной классификации |
| **Скрытые слои** | Зависит от решаемой задачи, от одного и больше | Аналогично бинарной классификации |
| **Количество нейронов в скрытом слое** | Зависит от задачи, обычно от 10 до 512 | Аналогично бинарной классификации |
| **Количество нейронов в выходном слое** (`out_features`) | 1 (предсказывает относится ли объект к некоторому классу или нет) | 1 на класс (если объект может быть отнесен к 1 из 10 классов, потребуется 10 нейронов) |
| **Функция активации стрытого слоя** | Обычно [ReLU](https://pytorch.org/docs/stable/generated/torch.nn.ReLU.html#torch.nn.ReLU) (rectified linear unit) но [может быть и другой](https://en.wikipedia.org/wiki/Activation_function#Table_of_activation_functions) | Аналогично бинарной классификации |
| **Функция активации выходного слоя** | [Sigmoid](https://en.wikipedia.org/wiki/Sigmoid_function) ([`torch.sigmoid`](https://pytorch.org/docs/stable/generated/torch.sigmoid.html) в PyTorch)| [Softmax](https://en.wikipedia.org/wiki/Softmax_function) ([`torch.softmax`](https://pytorch.org/docs/stable/generated/torch.nn.Softmax.html) в PyTorch) |
| **Функция потерь** | [Binary crossentropy](https://en.wikipedia.org/wiki/Cross_entropy#Cross-entropy_loss_function_and_logistic_regression) ([`torch.nn.BCELoss`](https://pytorch.org/docs/stable/generated/torch.nn.BCELoss.html) в PyTorch) | Cross entropy ([`torch.nn.CrossEntropyLoss`](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html) в PyTorch) |
| **Оптимизатор** | [SGD](https://pytorch.org/docs/stable/generated/torch.optim.SGD.html) (стохастический градиентный спуск), [Adam](https://pytorch.org/docs/stable/generated/torch.optim.Adam.html) (см. [`torch.optim`](https://pytorch.org/docs/stable/optim.html)) | Аналогично бинарной классификации |

# Решение задачи бинарной классификации

## Генерация набора данных



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_circles


n_samples = 500

X, y = make_circles(n_samples,
                    noise=0.03, 
                    random_state=0)

plt.figure(figsize=(6, 6))

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
_ = plt.scatter(X[:, 0], X[:, 1], c=y, cmap="Set3", ec='black', alpha=0.85, lw=0.5)
```

```python
print(X.shape, y.shape)
```

Количество признаков, которым характеризуется точка на плоскости равняется двум. В качестве метки класса выступает значение 0 или 1 (к какой группе точек относится та или иная точка). Следовательно в первом слое нейронной сети будет 2 нейрона.

## Разделим выборку на тренировочную и тестовую

```python
import torch

X = torch.from_numpy(X).type(torch.float32)
y = torch.from_numpy(y).type(torch.float32)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
def train_test_plot(X_train, y_train, X_test, y_test, title1="Train", title2="Test"):
    plt.figure(figsize=(16, 7))

    ax = plt.subplot(121)
    plt.title(title1)
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    _ = plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap="Set3", ec='black', alpha=0.85, lw=0.5)

    plt.subplot(122, sharex=ax, sharey=ax)
    plt.title(title2)
    plt.xlabel("$x_1$")
    # Если классификатор предсказал вероятность принадлежности первому классу >= 0.5, считаем объект принадлежащим классу 1
    # в противном случае - 0
    _ = plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap="Set3", ec='black', alpha=0.85, lw=0.5)

train_test_plot(X_train, y_train, X_test, y_test)
```

 ## Построение модели
 
1. Создаем класс, который является наследником `nn.Module`.
2. Задать функцию потерь и оптимизатор.
3. Производим обучение модели.

```python
# 1. Создаваемый класс обязан быть наследником класса nn.Module
class CircleModelV0(torch.nn.Module):
    def __init__(self):
        super().__init__()
        # 2. Входной слой содержит 2 нейрона (так как объект описывается двумя признаками)
        # Скрытый слой содержит 5 нейронов
        # Выходной слой содержит 1 нейрон, который говорит принадлежит ли объект классу 1 или нет
        self.layer_1 = torch.nn.Linear(in_features=2, out_features=5) 
        self.layer_2 = torch.nn.Linear(in_features=5, out_features=1)
    
    # 3. Определим функцию для вычисления метки по объекту
    def forward(self, x):
        return self.layer_2(self.layer_1(x))

# 4. Создадим объект модели
model = CircleModelV0()
model
```

Выполним предсказания для необученной модели:

```python
y_pred = model(X_test)
print(f"Количество предсказаний: {len(y_pred)}, Shape: {y_pred.shape}")
print(f"Количество меток в тестовой выборке: {len(y_test)}, Shape: {y_test.shape} \n")
print(f"Первые 10 предсказаний:\n{y_pred[:10]} \n")
print(f"Первые 10 метод из тестовой выборки:\n{y_test[:10]}")
```

```python
train_test_plot(X_test, y_test, X_test, y_pred >= 0.5, "Test", "Predictions")
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(in_features=2, out_features=5),
    torch.nn.Linear(in_features=5, out_features=1)
)
```

## Выбор функции потерь и оптимизатора

В зависимости от решаемой задачи вы должны выбирать правильную функцию потерь. Например для регресси можно использовать mean absolute error (MAE). Для бинарной классификации часто используется [binary cross entropy](https://towardsdatascience.com/understanding-binary-cross-entropy-log-loss-a-visual-explanation-a3ac6025181a) в качестве функции потерь.

В качестве оптимизатора может использоваться стохастических градиентный спуск. (SGD, `torch.optim.SGD()`) и Adam optimizer (`torch.optim.Adam()`). 

| Функция потерь / Оптимизатор | Задача | Код |
| ----- | ----- | ----- |
| Стохастический градиентный спуск (SGD) | Классификация, регрессия и другие | [`torch.optim.SGD()`](https://pytorch.org/docs/stable/generated/torch.optim.SGD.html) |
| Adam Optimizer | Классификация, регрессия и другие | [`torch.optim.Adam()`](https://pytorch.org/docs/stable/generated/torch.optim.Adam.html) |
| Binary cross entropy loss | Бинарная классификация | [`torch.nn.BCELossWithLogits`](https://pytorch.org/docs/stable/generated/torch.nn.BCEWithLogitsLoss.html) или [`torch.nn.BCELoss`](https://pytorch.org/docs/stable/generated/torch.nn.BCELoss.html) |
| Cross entropy loss | Многоклассовая классификация | [`torch.nn.CrossEntropyLoss`](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html) |
| Mean absolute error (MAE) или L1 Loss | Регрессия | [`torch.nn.L1Loss`](https://pytorch.org/docs/stable/generated/torch.nn.L1Loss.html) | 
| Mean squared error (MSE) или L2 Loss | Регрессия | [`torch.nn.MSELoss`](https://pytorch.org/docs/stable/generated/torch.nn.MSELoss.html#torch.nn.MSELoss) |  

```python
loss_fn = torch.nn.BCEWithLogitsLoss()

optimizer = torch.optim.SGD(params=model.parameters(), 
                            lr=0.1)
```

# Обучение модели

```python
# Вычисление точности
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item() # torch.eq() calculates where two tensors are equal
    acc = (correct / len(y_pred)) * 100 
    return acc
```

```python
def learn_binary_classification_model(model, n_epochs, print_every_epoch=None):
    if print_every_epoch is None:
        print_every_epoch = n_epochs / 10
    
    torch.manual_seed(42)

    for epoch in range(n_epochs):
        ### Обучение
        model.train()

        # 1. Получаем предсказания
        y_logits = model(X_train).squeeze()
        y_pred = torch.round(torch.sigmoid(y_logits)) # получаем предсказание (0 или 1)

        # 2. Вычисление функции потерь и точности (accuracy)
        loss = loss_fn(y_logits,
                       y_train) 
        acc = accuracy_fn(y_true=y_train, 
                          y_pred=y_pred) 

        # 3. Очистка градиентов
        optimizer.zero_grad()

        # 4. Вычисление градиента
        loss.backward()

        # 5. Очередной шаг спуска
        optimizer.step()

        ### Тестирование
        model.eval()
        with torch.inference_mode():
            # 1. Вычисление значений на тесте
            test_logits = model(X_test).squeeze() 
            test_pred = torch.round(torch.sigmoid(test_logits))
            # 2. Вычисление функции потерь и точности (accuracy)
            test_loss = loss_fn(test_logits,
                                y_test)
            test_acc = accuracy_fn(y_true=y_test,
                                   y_pred=test_pred)

        # Выводим статистику каждые 10 итераций
        if epoch % print_every_epoch == 0:
            print(f"Epoch: {epoch:3} | Loss: {loss:.5f}, Accuracy: {acc:.2f}% | Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
    
    
learn_binary_classification_model(model, 100)
```

Согласно полученным логам, наш классификатор будет в 50% случаев определять класс верно.

```python
import numpy as np

def plot_decision_boundary(model: torch.nn.Module, X: torch.Tensor, y: torch.Tensor):
    """Plots decision boundaries of model predicting on X in comparison to y."""
    model.to("cpu")
    X, y = X.to("cpu"), y.to("cpu")

    margin_x1 = (X[:, 0].max() - X[:, 0].min()) * 0.1
    margin_x2 = (X[:, 1].max() - X[:, 1].min()) * 0.1
    
    x_min, x_max = X[:, 0].min() - margin_x1, X[:, 0].max() + margin_x1
    y_min, y_max = X[:, 1].min() - margin_x2, X[:, 1].max() + margin_x2
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 501), np.linspace(y_min, y_max, 501))

    X_to_pred_on = torch.from_numpy(np.column_stack((xx.ravel(), yy.ravel()))).float()

    model.eval()
    with torch.inference_mode():
        y_logits = model(X_to_pred_on)

    if len(torch.unique(y)) > 2:
        y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1)
    else:
        y_pred = torch.round(torch.sigmoid(y_logits))

    # Reshape preds and plot
    y_pred = y_pred.reshape(xx.shape).detach().numpy()
    plt.contourf(xx, yy, y_pred, cmap="Set3", alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap="Set3", ec='black', alpha=1, lw=0.5)
    
    plt.xlabel("$x_0$")
    plt.ylabel("$x_1$")
    
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

    
def train_test_model_plot(model, X_train, y_train, X_test, y_test):
    plt.figure(figsize=(16, 7))

    plt.subplot(1, 2, 1)
    plt.title("Train")
    plot_decision_boundary(model, X_train, y_train)

    plt.subplot(1, 2, 2)
    plt.title("Test")
    plot_decision_boundary(model, X_test, y_test)
    
    
train_test_model_plot(model, X_train, y_train, X_test, y_test)
```

Наша модель не объясняет данные, модель слишком простая (в данных более сложная зависимость). Такая ситуация называется **недообучением**. На самом деле какую бы мы не построили нейронную сеть без нелинейных функций активации, граница разделения будет линейной.

# Способы решения проблемы недообучения

| Техника* | Как это работает? |
| :-----: | ----- |
| **Добавить больше слоев** | Каждый слой *потенциально* увеличивает возможности обучения модели, поскольку каждый слой может изучать какой-то новый шаблон в данных |
| **Добавить больше нейронов в скрытый слой** | Идея аналогично пункту выше. |
| **Увеличить количество эпох** | Большее количество итераций обучения позволяет лучше исследовать закономерности в данных |
| **Изменить функцию активации** | Некоторые данные не могут быть линейно разделены. Требуется использование некоторых нелинейных функций активаций |
| **Изменить скорость обучения** | Неправильно подобранная скорость обучения может приводить к проблемам недообучения |
| **Изменить функцию потерь** | Неверно подобранная функция потерь может не привести к корректному решению задачи. |
| **Использовать предобученные модели** | Некоторые, возможно, решали похожие задачи. Возможно, стоит использовать предобученную модель и при необходимости дообучить |

# Решение задачи мультиклассовой классификации

```python
from sklearn.datasets import make_blobs

NUM_CLASSES = 5
NUM_FEATURES = 2
RANDOM_SEED = 6

# Создание набора данных
X_blob, y_blob = make_blobs(n_samples=1000,
    n_features=NUM_FEATURES,
    centers=NUM_CLASSES,
    cluster_std=1.5,
    random_state=RANDOM_SEED
)

# Преобразование в тензоры
X_blob = torch.from_numpy(X_blob).type(torch.float)
y_blob = torch.from_numpy(y_blob).type(torch.LongTensor)
print(X_blob[:5], y_blob[:5])
```

```python
# Разбиение на тренировочную и тестовую выборку
X_blob_train, X_blob_test, y_blob_train, y_blob_test = train_test_split(X_blob,
    y_blob,
    test_size=0.2,
    random_state=RANDOM_SEED
)

train_test_plot(X_blob_train, y_blob_train, X_blob_test, y_blob_test)
```

```python
class MulticlassClassification(torch.nn.Module):
    def __init__(self, input_features, output_features, hidden_units=8):
        super().__init__()
        self.linear_layer_stack = torch.nn.Sequential(
            torch.nn.Linear(in_features=input_features, out_features=hidden_units),
            torch.nn.Linear(in_features=hidden_units, out_features=hidden_units),
            torch.nn.Linear(in_features=hidden_units, out_features=output_features),
        )
    
    def forward(self, x):
        return self.linear_layer_stack(x)

    
model = MulticlassClassification(input_features=NUM_FEATURES, 
                                 output_features=NUM_CLASSES, 
                                 hidden_units=8)
model
```

```python
# Опишем функцию потерь и оптимизатор
loss_fn = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), 
                            lr=0.1) # Попробуйте различные learning_rate
```

Посмотрим на предсказания необученной модели:

```python
train_test_model_plot(model, X_blob_train, y_blob_train, 
                             X_blob_test, y_blob_test)
```

```python
def learn_milticlass_classification_model(model, n_epochs, print_every_epoch=None):
    if print_every_epoch is None:
        print_every_epoch = n_epochs / 10
    
    torch.manual_seed(42)

    for epoch in range(n_epochs):
        ### Обучение
        model.train()

        y_logits = model(X_blob_train)
        y_pred = torch.softmax(y_logits, dim=1).argmax(dim=1)
        
        loss = loss_fn(y_logits, y_blob_train) 
        acc = accuracy_fn(y_true=y_blob_train,
                          y_pred=y_pred)

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        ### Тестирование
        model.eval()
        with torch.inference_mode():
            test_logits = model(X_blob_test)
            test_pred = torch.softmax(test_logits, dim=1).argmax(dim=1)
            
            test_loss = loss_fn(test_logits, y_blob_test)
            test_acc = accuracy_fn(y_true=y_blob_test,
                                 y_pred=test_pred)

        # Print out what's happening
        if epoch % print_every_epoch == 0:
            print(f"Epoch: {epoch} | Loss: {loss:.5f}, Acc: {acc:.2f}% | Test Loss: {test_loss:.5f}, Test Acc: {test_acc:.2f}%") 
            
learn_milticlass_classification_model(model, 500, 50)
```

   * Добейтесь качества хотя бы в 96% на тесте для случая с нелинейными преобразованиями
   * Прогресс обучения должен выводиться в каждые 10 итераций

```python
from torchmetrics import Accuracy

def learn_binary_classification_model(model, optimizer, n_epochs, print_every_epoch=None):
    if print_every_epoch is None:
        print_every_epoch = n_epochs / 10
    
    torch.manual_seed(42)

    for epoch in range(n_epochs):
        ### Обучение
        model.train()

        y_logits = model(X_train).squeeze()
        y_pred = torch.round(torch.sigmoid(y_logits))

        loss = loss_fn(y_logits, y_train) 
        acc = Accuracy(task="binary")(y_train, y_pred)
    
        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        ### Тестирование
        model.eval()
        with torch.inference_mode():
            test_logits = model(X_test).squeeze() 
            test_pred = torch.round(torch.sigmoid(test_logits))
            
            test_loss = loss_fn(test_logits, y_test)
            test_acc = Accuracy(task="binary")(y_test, test_pred)

        if epoch % print_every_epoch == 0:
            print(f"Epoch: {epoch:3} | Loss: {loss:.5f}, Accuracy: {acc:.2f}% | Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
```

5. Выполните многоклассовую классификацию датасета ниже [spirals data creation function from CS231n](https://cs231n.github.io/neural-networks-case-study/):

```python
N = 300 
D = 2 
K = 3
X = np.zeros((N*K,D)) 
y = np.zeros(N*K, dtype='uint8') 
for j in range(K):
    ix = range(N*j,N*(j+1))
    r = np.linspace(0.0, 1, N) # radius
    t = np.linspace(j*4, (j+1)*4, N) + np.random.randn(N)*0.2 # theta
    X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
    y[ix] = j
```

Требования:
   * Модель может содержать и линейные и нелинейные слои.
   * В качестве оптимизатора используйте Adam optimizer вместо SGD.
   * Добейтесь точности в 95%.
   * Визуализируйте результаты.
