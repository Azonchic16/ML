```python
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.optim as optim
import torch
```

```python
x = np.linspace(-2.5, 2.5, 300)
x_reshaped = x.reshape(-1, 1)
y = np.sin(np.power(x_reshaped, 3))

x_tensor = torch.tensor(x_reshaped, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)
```

```python
class MBGD(nn.Module):
    def __init__(self):
        super(MBGD, self).__init__()
        self.fc1 = nn.Linear(1, 50)
        self.fc2 = nn.Linear(50, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
mbgd = MBGD()
criterion = nn.MSELoss()
optimizer = optim.SGD(mbgd.parameters(), lr=0.01)
```

```python
batch_size = 32
loss_arr = []

for epoch in range(10000):
    indices = torch.randperm(x_tensor.size(0))
    for start_index in range(0, x_tensor.size(0), batch_size):
        end_index = min(start_index + batch_size, x_tensor.size(0))
        batch_indices = indices[start_index:end_index]
        batch_x, batch_y = x_tensor[batch_indices], y_tensor[batch_indices]

        optimizer.zero_grad()
        output = mbgd(batch_x)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()
        loss_arr.append(loss.item())
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/5000], Loss: {loss.item():.4f}')
```

```python
with torch.no_grad():
    predicted = mbgd(x_tensor).numpy()

fig, ax = plt.subplots(2, 1)
ax[1].plot(np.arange(1, len(loss_arr) + 1), loss_arr)
ax[0].plot(x, y, label='Исходные данные', color='blue')
ax[0].plot(x, predicted, color='red', label='Предсказанные значения')
ax[0].legend()
plt.show()
```
